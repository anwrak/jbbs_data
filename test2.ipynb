```python
import numpy as np
import pandas as pd
from pybaseball import team_pitching
from plotnine import ggplot, geom_line, aes, theme_light, geom_point
```

```python
pitching_data = team_pitching(2013, 2023)
```

```python
yearly_k = pitching_data[['Season', 'SO', 'HR']]
yearly_k

(
    ggplot(yearly_k[yearly_k['Season'] != 2020], aes(x='HR', y='SO'))
    + geom_point()
    + theme_light()
)
```
